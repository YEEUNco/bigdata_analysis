```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/5_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/5_test.csv')

display(train)
display(test)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

train_pre = train.copy()
train_pre = train_pre.drop(columns=['ID'])
train_pre = train_pre[~train_pre['model'].isin(['230','180','200'])]
train_pre = train_pre[~train_pre['fuelType'].isin(['Other'])]
train_pre = train_pre[~train_pre['transmission'].isin(['Other'])]
sc = StandardScaler()
sc.fit(train_pre[['year', 'mileage', 'tax', 'mpg']])
train_pre[['year', 'mileage', 'tax', 'mpg']] = sc.transform(train_pre[['year', 'mileage', 'tax', 'mpg']])

le = LabelEncoder()
le.fit(train_pre['model'])
train_pre['model'] = le.transform(train_pre['model'])

train_pre = pd.get_dummies(train_pre, drop_first=True)
train_pre.info()

x=train_pre.drop(columns=['price'])
y=train_pre['price']
```

```python
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_train_pred = rf.predict(X_train)

test_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
train_rmse = np.sqrt(mean_squared_error(y_train, rf_train_pred))
print(test_rmse)
print(train_rmse)
```
