```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
#결측치 처리
#enrolled_university, education_level, experience, last_new_job 정도는 최빈값으로 대체하면 될 듯
#gender, major_discipline, company_size, company_type 은 어떻게 해야할까?

x_train['enrolled_university'] = x_train['enrolled_university'].fillna(x_train['enrolled_university'].value_counts().index[0])
x_train['education_level'] = x_train['education_level'].fillna(x_train['education_level'].value_counts().index[0])
x_train['experience'] = x_train['experience'].fillna(x_train['experience'].value_counts().index[0])
x_train['last_new_job'] = x_train['last_new_job'].fillna(x_train['last_new_job'].value_counts().index[0])
```

```python
x_train['gender'] = x_train['gender'].fillna('unknown')
x_train['major_discipline'] = x_train['major_discipline'].fillna('unknown')
x_train['company_type'] = x_train['company_type'].fillna('unknown')
x_train['company_size'] = x_train['company_size'].fillna('unknown')
```

```python
x_train.isnull().sum()

```

```python
for col in x_train.select_dtypes(include='object').columns:
  print(x_train[col].value_counts())
  print('\n''\n')
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

sc = StandardScaler()
sc.fit(x_train[['training_hours']])
x_train[['training_hours']] = sc.transform(x_train[['training_hours']])
x_train.describe()
```

```python
x_train = x_train.drop(columns=['enrollee_id'])
y_train = y_train.drop(columns=['enrollee_id'])
```

```python
#label을 할 것인지 get_dummies를 할 것인지..
#label
# experience
#get_dummies
#gender, relevent_exp, enrolled_university, education_level, experience빼고 다 해도 될 듯?
```

```python
le = LabelEncoder()
le.fit(x_train['experience'])
x_train['experience'] = le.transform(x_train['experience'])
le2 = LabelEncoder()
le2.fit(x_train['company_size'])
x_train['company_size'] = le2.transform(x_train['company_size'])
```

```python
x_train = pd.get_dummies(
    x_train,
    columns=['gender', 'relevent_experience', 'enrolled_university',
             'education_level', 'major_discipline', 'company_type', 'last_new_job'],
    drop_first=True
)
```

```python
x_train = x_train.drop(columns=['city'])
```

```python
x_train.info()
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(x_train, y_train, test_size=0.2, random_state=0, stratify=y_train)
```

```python
x_tr.shape
y_tr.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0,max_depth=15)
rf.fit(x_tr, y_tr)
rf_pred = rf.predict(x_te)
rf_pred
rf_pred_prob = rf.predict_proba(x_te)
rf_pred_prob
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_te, rf_pred)
score
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score
score = accuracy_score(y_te, rf_pred)
score
```

```python
score = roc_auc_score(y_te, rf_pred_prob[:,1])
score
```
