```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}
df['newIncome'] = df.Income_Category.map(lambda x:dic[x])
df['newIncome']
```

```python
df.head()
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df['newIncome'] = df.Income_Category.apply(changeCategory)
df.head()
```

```python
df['AgeState'] = df.Customer_Age.apply(lambda x: x//10)
df.AgeState.head()

df.groupby('AgeState').size()
```

```python
df['newEduLevel'] = df.Education_Level.apply(lambda x: 1 if 'Graduate' in x else 0)
df.newEduLevel.head()
df.groupby('newEduLevel').size()
df.['newLimit'].value_counts()
```

```python
df['newLimit'] = df.Credit_Limit.apply(lambda x: 1 if x >= 4500 else 0)
df.groupby('newLimit').size()
```

```python
df['newState']= df.apply(lambda x: 1 if (x.Marital_Status == 'Married')and(x.Card_Category=='Platinum') else 0 ,
                        axis=1)
df['newState'].value_counts()
```

```python
def check(x):
    if (x.Marital_Status == 'Married') and ( x.Card_Category == 'Platinum'):
        return 1
    else:
        return 0
df['newState'] = df.apply(check,axis=1)
df['newState'].value_counts()
```

```python
df['Gender'] = df.Gender.apply(lambda x: 'Male' if x=='M' else 'Female')
df.head()
```
