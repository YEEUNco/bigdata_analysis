```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/6_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/6_test.csv')

display(train)
display(test)
```

```python
train_pre = train.copy()
train_pre = train_pre.drop(columns=['ID'])
train_pre = train_pre[~train_pre['Age_Category'].isnull()]
train_pre['Skin_Cancer'] = train_pre['Skin_Cancer'].fillna(train_pre['Skin_Cancer'].value_counts().index[0])
train_pre.isnull().sum()
```

```python
for col in train_pre.select_dtypes(include='object').columns:
  print(train_pre[col].value_counts())
  print('\n\n')
```

```python
import numpy as np
print(train.describe().T[['mean', 'std', 'min', '50%', 'max']])

for col in train.select_dtypes(exclude='object').columns:
  print(col)
  max_iqr = (train[col].quantile(0.75)-train[col].quantile(0.25))*1.5+train[col].quantile(0.75)
  print((train[col].quantile(0.75)-train[col].quantile(0.25))*1.5+train[col].quantile(0.75))
  print(train[train[col]>max_iqr].shape[0])
  print('\n')

print(train.shape)
print(train.select_dtypes(exclude='object').columns)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split


sc = StandardScaler()
sc.fit(train_pre[['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']])
train_pre[['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']] = sc.transform(train_pre[['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']])
train_pre.describe().T

le = LabelEncoder()
le.fit(train_pre['Age_Category'])
train_pre['Age_Category'] = le.transform(train_pre['Age_Category'])

x=train_pre.drop(columns=['General_Health'])
y=train_pre['General_Health']

x = pd.get_dummies(x,drop_first=True)
x = x.drop(columns=['Checkup_Never'])
print(x.shape)
print(x.info())
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0,stratify=y)

rf = RandomForestClassifier(random_state=0,max_depth=8)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_train_pred = rf.predict(X_train)

print(f1_score(y_test, rf_pred, average='micro'))
print(f1_score(y_train, rf_train_pred, average='micro'))
print(rf.classes_)
```

```python
test_pre = test.copy()
test_pre = test_pre.drop(columns=['ID'])
test_pre['Age_Category'] = le.transform(test_pre['Age_Category'])
test_pre[['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']] = sc.transform(test_pre[['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']])

test_pre['BMI'] = test_pre['BMI'].fillna(test_pre['Weight_(kg)']/((test_pre['Height_(cm)']/100)**2))

test_pre = pd.get_dummies(test_pre,drop_first=True)
test_pre = test_pre[list(x.columns)]
print(test_pre.info())
print(x.info())
test_pre.describe().T
```

```python
test_pred = rf.predict(test_pre)
test_pred

df = pd.DataFrame(test_pred, columns=['pred'])
df
```
