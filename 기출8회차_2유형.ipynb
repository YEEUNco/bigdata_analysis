```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/8_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/8_test.csv')

display(train)
display(test)
```

```python
train.info()
train.describe(include='O')
#train.isnull().sum()
train.describe().T
```

```python
test.info()
test.describe(include='O')
test.isnull().sum()
```

```python
train_pre = train.copy()
train_pre['Date'] = pd.to_datetime(train_pre['Date'])
train_pre['month'] = train_pre['Date'].dt.month
train_pre = train_pre[~train_pre['C4'].isnull()]
train_pre['C1'] = train_pre['C1'].astype('str').str.replace('X_', "").astype(int)
train_pre['C2'] = train_pre['C2'].astype('str').str.replace('Y_', '').astype(int)
train_pre['C3'] = train_pre['C3'].astype('str').str.replace('Z_', '').astype(int)
train_pre['C4'] = train_pre['C4'].astype('str').str.replace('W_', '').astype(int)
train_pre['Pressure'] = train_pre['Pressure'].fillna(train_pre['Pressure'].mean())
x = train_pre.drop(columns=['Date','Y'])
y = train_pre['Y']

train_pre.info()


test_pre = test.copy()
test_pre['Date'] = pd.to_datetime(test_pre['Date'])
test_pre['month'] = test_pre['Date'].dt.month
test_pre['C1'] = test_pre['C1'].fillna(test_pre['C1'].value_counts().index[0])
test_pre['C1'] = test_pre['C1'].astype('str').str.replace('X_', "").astype(int)
test_pre['C2'] = test_pre['C2'].astype('str').str.replace('Y_', '').astype(int)
test_pre['C3'] = test_pre['C3'].astype('str').str.replace('Z_', '').astype(int)
test_pre['C4'] = test_pre['C4'].astype('str').str.replace('W_', '').astype(int)
test_pre['Pressure'] = test_pre['Pressure'].fillna(test_pre['Pressure'].mean())
test_pre['Precipitation'] = test_pre['Precipitation'].fillna(test_pre['Precipitation'].mean())
test_pre.info()
test_pre = test_pre.drop(columns=['Date'])
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_train_pred = rf.predict(X_train)

print(mean_absolute_error(y_test, rf_pred))
print(mean_absolute_error(y_train, rf_train_pred))
print()

xgb=XGBRegressor(random_state=0)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
xgb_train_pred = xgb.predict(X_train)

print(mean_absolute_error(y_test, xgb_pred))
print(mean_absolute_error(y_train, xgb_train_pred))
```
