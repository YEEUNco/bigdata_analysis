```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/10/t2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/10/t2_test.csv')

display(train.sort_values('작업_시작_일자'))
display(test.head(2))
```

```python
train.info()
for col in train.select_dtypes(include='object').columns:
  print(train[col].value_counts())
  print('\n')

train.describe().T
#train['작업_연도'].value_counts()
```

```python
test.info()
test.describe()
```

```python
#작업시작일자 -> month
#등급 -> labelEncoder

#나머진 get_dummies
#작업_연도는 drop
#StandardScaler

from sklearn.preprocessing import StandardScaler

train_pre = train.copy()
train_pre['작업_시작_일자'] = pd.to_datetime(train_pre['작업_시작_일자'])
train_pre['month'] = train_pre['작업_시작_일자'].dt.month
grade = {
    'A':3,
    'B':2,
    'C':3
}
train_pre['등급'] = train_pre['등급'].apply(lambda x: grade[x])
train_pre = train_pre.drop(columns=['작업_시작_일자'])

sc = StandardScaler()
sc.fit(train_pre[['작업_소요_일수', '재배면적', '평균_기온', '작업자_수']])
train_pre[['작업_소요_일수', '재배면적', '평균_기온', '작업자_수']] = sc.transform(train_pre[['작업_소요_일수', '재배면적', '평균_기온', '작업자_수']])

train_pre = pd.get_dummies(train_pre, drop_first=True)

train_pre.info()
train_pre.shape

x= train_pre.drop(columns=['작업_유형', '작업_연도'])
y=train_pre['작업_유형']

test_pre = test.copy()
test_pre['작업_시작_일자'] = pd.to_datetime(test_pre['작업_시작_일자'])
test_pre['month'] = test_pre['작업_시작_일자'].dt.month
test_pre['등급'] = test_pre['등급'].apply(lambda x: grade[x])
test_pre = test_pre.drop(columns=['작업_시작_일자', '작업_연도'])
test_pre[['작업_소요_일수', '재배면적', '평균_기온', '작업자_수']] = sc.transform(test_pre[['작업_소요_일수', '재배면적', '평균_기온', '작업자_수']])
test_pre = pd.get_dummies(test_pre, drop_first=True)

test_pre.info()
test_pre.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from xgboost import XGBClassifier

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0, stratify=y)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_train_pred = rf.predict(X_train)
print(rf.classes_)

print(f1_score(y_test, rf_pred, average='micro'))
print(f1_score(y_train, rf_train_pred, average='micro'))
print()

xgb= XGBClassifier(random_state=0)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
xgb_train_pred = xgb.predict(X_train)
print(f1_score(y_test, xgb_pred, average='micro'))
print(f1_score(y_train, xgb_train_pred, average='micro'))
```

```python
test_pred = rf.predict(test_pre)
df = pd.DataFrame(test_pred, columns=['pred'])
df
```
