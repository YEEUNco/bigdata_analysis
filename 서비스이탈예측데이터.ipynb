```python
import pandas as pd

X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
X_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")
```

```python
for col in X_train.select_dtypes(include='object'):
  print(X_train[col].value_counts())
  print('\n''\n')
```

```python
X_train = X_train.drop(columns=['CustomerId', 'Surname'])
X_train.describe()
```

```python
X_train['Gender'] = X_train['Gender'].str.lower()
X_train['Gender'] = X_train['Gender'].str.replace(" ", '')
X_train['Gender'].value_counts()
#X_train= pd.get_dummies(X_train)
#X_train.head()
```

```python
X_train= pd.get_dummies(X_train)
X_train.head()
```

```python
#CreditScore, Age, Balance, EstimatedSalary에 대해서 sc
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train[['CreditScore', 'Age', 'Balance', 'EstimatedSalary']])
X_train.loc[:,['CreditScore', 'Age', 'Balance', 'EstimatedSalary']] = sc.transform(X_train[['CreditScore', 'Age', 'Balance', 'EstimatedSalary']])
X_train.describe()
```

```python
y_train = y_train.drop(columns=['CustomerId'])
```

```python
from sklearn.model_selection import train_test_split

x_tr, x_te, y_tr, y_te = train_test_split(X_train, y_train, test_size=0.2,random_state=0, stratify=y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier



#X_test = X_test.drop(columns=['CustomerId'])
#y_train = y_train.drop(columns=['CustomerId'])
rf = RandomForestClassifier(random_state=124)
rf.fit(x_tr, y_tr)
rf_pred = rf.predict(x_te)
```

```python
from sklearn.metrics import f1_score
f1_score(y_te, rf_pred)
```
