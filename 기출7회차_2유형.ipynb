```python
train_pre = train.copy()
train_pre = train_pre.drop(columns=['ID'])
train_pre = train_pre[~train_pre['업종명'].isnull()]
y = train_pre['이용금액']

#train_pre['연월'] = pd.to_datetime(train_pre['연월'])
train_pre['이용건수'] = train_pre['이용건수'].fillna(train_pre['이용건수'].mean())
train_pre['월'] = train_pre['연월'].astype('str').str[-2:].astype(int)
train_pre = pd.get_dummies(train_pre, drop_first=True)
train_pre = train_pre.drop(columns=['연월'])

test_pre = test.copy()
test_pre['이용건수'] = test_pre['이용건수'].fillna(test_pre['이용건수'].mean())
test_pre['월'] = test_pre['연월'].astype('str').str[-2:].astype(int)
test_pre = test_pre.drop(columns=['연월','ID'])
test_pre = pd.get_dummies(test_pre, drop_first=True)


train_pre.info()
print(test_pre.info())

print(train_pre.shape)
print(test_pre.shape)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

x = train_pre.drop(columns=['이용금액'])
y = train_pre['이용금액']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0, max_depth=20)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_train_pred = rf.predict(X_train)

print(np.sqrt(mean_squared_error(y_test, rf_pred)))
print(np.sqrt(mean_squared_error(y_train, rf_train_pred)))
print()

xgb = XGBRegressor(random_state=0, max_depth=16)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
xgb_train_pred = xgb.predict(X_train)
print(np.sqrt(mean_squared_error(y_test, xgb_pred)))
print(np.sqrt(mean_squared_error(y_train, xgb_train_pred)))
print(xgb_pred)
```

```python
test_pred = xgb.predict(test_pre)
result = pd.DataFrame(test_pred, columns=['pred'])
result
```
