```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/x_test.csv")
y_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/y_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
x_train = x_train.drop(columns=['ID', '구강검진수검여부'])
```

```python
x_train.describe().T
```

```python
from sklearn.preprocessing import StandardScaler

#연령대코드(5세단위), 신장(5Cm단위), 체중(5Kg단위), 허리둘레, 수축기혈압, 이완기혈압, 식전혈당(공복혈당), 총콜레스테롤, 트리글리세라이드, HDL콜레스테롤	, LDL콜레스테롤, 혈색소, (혈청지오티)AST, (혈청지오티)ALT, 감마지티피
sc = StandardScaler()
sc.fit(x_train[['연령대코드(5세단위)', '신장(5Cm단위)', '체중(5Kg단위)', '허리둘레', '수축기혈압', '이완기혈압', '식전혈당(공복혈당)', '총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤', '혈색소', '(혈청지오티)AST', '(혈청지오티)ALT', '감마지티피']])
x_train[['연령대코드(5세단위)', '신장(5Cm단위)', '체중(5Kg단위)', '허리둘레', '수축기혈압', '이완기혈압', '식전혈당(공복혈당)', '총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤', '혈색소', '(혈청지오티)AST', '(혈청지오티)ALT', '감마지티피']] = sc.transform(x_train[['연령대코드(5세단위)', '신장(5Cm단위)', '체중(5Kg단위)', '허리둘레', '수축기혈압', '이완기혈압', '식전혈당(공복혈당)', '총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤', '혈색소', '(혈청지오티)AST', '(혈청지오티)ALT', '감마지티피']])
x_train.describe().T
```

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_train.info()
```

```python
y_train.info()
```

```python
y_train = y_train.drop(columns=['ID'])
```

```python
y_train.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
from sklearn.model_selection import train_test_split

x_tr, x_te, y_tr, y_te = train_test_split(x_train, y_train, test_size=0.2, random_state=0, stratify=y_train)

rf = RandomForestClassifier()
rf.fit(x_tr, y_tr)
rf_pred = rf.predict(x_te)
rf_pred_prob = rf.predict_proba(x_te)
print(rf_pred)
print(rf_pred_prob)
```

```python
f1 = f1_score(y_te, rf_pred)
roc_auc = roc_auc_score(y_te, rf_pred_prob[:,1])
accuracy = accuracy_score(y_te, rf_pred)
print(f1)
print(roc_auc)
print(accuracy)
```
