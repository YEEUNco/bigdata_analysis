```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/4_train.csv', index_col=0)
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/4_test.csv', index_col=0)

display(train)
display(test)
```

```python
x = train.drop(columns=['ID','Segmentation'])
y = train['Segmentation']
```

```python
x.info()
```

```python
x.isnull().sum()
```

```python
x.describe()
```

```python
for col in x.select_dtypes(include='object').columns:
  print(x[col].value_counts())
  print('\n\n')
```

```python
x=pd.get_dummies(x,drop_first=True)
x.info()
```

```python
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)

rf = RandomForestClassifier(max_depth=10)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_pred_prob = rf.predict_proba(X_test)
#print(rf_pred)
print(rf_pred_prob)
print(f1_score(y_test, rf_pred, average='macro'))
print(accuracy_score(y_test, rf_pred))
print('\n')

## xgb는 문자열 처리를 못한다 이걸 -> 수치형으로 바꿔서 판단할 수 있도록 해야하는 듯...
## 이래저래 randomforest가 짱이네..
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
print(f1_score(y_test, xgb_pred, average='macro'))
print(accuracy_score(y_test, xgb_pred))
```

```python
isn = test.isnull().sum()
isn = isn[isn!=0]
isn.index
```

```python
for col in isn.index:
  print(test[col].value_counts())
  print('\n\n')
```

```python
# ['Ever_Married', 'Graduated', 'Profession', 'Work_Experience', 'Family_Size', 'Var_1']

test_pre = test.copy()
test_pre['Ever_Married'] = test_pre['Ever_Married'].fillna(test_pre['Ever_Married'].value_counts().index[0])
test_pre['Graduated'] = test_pre['Graduated'].fillna(test_pre['Graduated'].value_counts().index[0])
test_pre['Profession'] = test_pre['Profession'].fillna(test_pre['Profession'].value_counts().index[0])
test_pre['Work_Experience'] = test_pre['Work_Experience'].fillna(test_pre['Work_Experience'].mean())
# ['Ever_Married', 'Graduated', 'Profession', 'Work_Experience', 'Family_Size', 'Var_1']
test_pre['Family_Size'] = test_pre['Family_Size'].fillna(test_pre['Family_Size'].mean())
test_pre['Var_1'] = test_pre['Var_1'].fillna(test_pre['Var_1'].value_counts().index[0])
test_pre = test_pre.drop(columns=['ID'])
test_pre = pd.get_dummies(test_pre)
```
